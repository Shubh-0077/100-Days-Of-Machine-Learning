```python
# nestd if-else
email = input("Enter Your Email:")
password  = input("Enter Your Password:")

if email == 'shubh@123gmail.com' and password == '123':
    print('welcome')
elif email == 'shubh@123gmail.com' and password != '123':
    print('Invalid password')
    password = input("Enter Your Password Again: ")
    if password == '123':
        print('Welcome Finally')
    else :
        print('Beta tumnse nah ho payega')
else: 
    print('Invalid email or password') 
```

```python
# min. of 3 nummbers

a = int(input('Enter first number: '))
b = int(input('Enter second number: '))
c = int(input('Enter third number: '))

if a<b and a<c :
    print('minimum is a ',a )
elif b<c :
    print('mimnum is b',b)
else :
    print('minimum is c ',c)
```

```python
# Menu driven Calulator 
fnum  = int(input('Enter first number: '))
snum  = int(input('Enter second number: '))

op = input('Enter your operator: ')

if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum * snum)
else:
    print(fnum / snum)
```

```python
# While

# TO print the Tabels (Basic eg)

num = int(input('Enter Your Number:'))

i = 1
while i < 11:
    print(num,'X',i,'=',num*i)
    i += 1
    
```

Question - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
cur_pop = 10000

for i in range(10,0,-1):
    print('yr:',i,'Population : ',cur_pop)
    cur_pop = cur_pop - 0.1*cur_pop
    
```

```python
# code
row = int(input('ENter Your Number :'))

for i in range(1,row+1):   # Loop for rows
    for j in range(1,i+1):   # Loop for printing * 
        print('*',end='')
    print()
```
