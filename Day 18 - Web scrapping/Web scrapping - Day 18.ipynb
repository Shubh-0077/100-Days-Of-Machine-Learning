# To consider the whole container

```python
company = soup.find_all('div' ,class_='companyCardWrapper')
```

```python
len(company)
```

```python
names= []
ratings = []
reviews = []
# ctype=[]
# hq=[]
# how_old=[]
# no_of_employee=[]

for i in company:
    names.append(i.find('h2').text.strip())
    ratings.append(i.find('div',class_='rating_text').text.strip())
    reviews.append(i.find('a',class_='companyCardWrapper__ActionWrapper').text.strip())
    
d = {'name': names,'ratings': ratings,'reviews':reviews}

df = pd.DataFrame(d)
```

Final database (getting all pages scrap)

```python
final = pd.DataFrame()

for j in range(1, 11):
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'} 
    webpage = requests.get(f'https://www.ambitionbox.com/list-of-companies?page={j}',headers=headers).text
    soup = BeautifulSoup(webpage, 'lxml')
    company = soup.find_all('div', class_='companyCardWrapper')

    names = []
    ratings = []
    reviews = []

    for i in company:
        names.append(i.find('h2').text.strip())

        rating_tag = i.find('div', class_='rating_text')
        ratings.append(rating_tag.text.strip() if rating_tag else None)

        review_tag = i.find('a', class_='companyCardWrapper__ActionWrapper')
        reviews.append(review_tag.text.strip() if review_tag else None)

    d = {'name': names, 'ratings': ratings, 'reviews': reviews}
    df = pd.DataFrame(d)

    final = pd.concat([final, df], ignore_index=True)
```
