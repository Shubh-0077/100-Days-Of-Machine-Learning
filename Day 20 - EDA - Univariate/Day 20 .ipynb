```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head()
```

Univariate (Single Column) is of 2 types :

### 1. Categorical Data

a. Count Plot

```python
# Using Pandas 
df['Sex'].value_counts().plot(kind='bar')
```

```python
# Using Seaborn
sns.countplot(df['Embarked'])
```

b. Pie

```python
df['Pclass'].value_counts().plot(kind='pie' ,autopct = '%.2F')
```

### 2. Numerical 

a. Histogram

```python
plt.hist(df['Age'], bins= 50)
```

b. Distplot

```python
sns.displot(df['Age'] ,kde = True)
```

c. BoxPlot

```python
sns.boxplot(df['Fare'])
```

```python
# Other basic Stats
print(' Min :',df['Age'].min())
print(' Max :',df['Age'].max())
print(' Skewness :',df['Age'].skew())  #(+ve : Right Skew , -ve : Left Skew)
print(' Mean ;',df['Age'].mean())
```
