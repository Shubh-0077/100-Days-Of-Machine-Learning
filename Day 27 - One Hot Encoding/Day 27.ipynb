# Using Scikit Learn 

Use **scikit-learn** because it keeps your encoding consistent when new data comes, while **pandas** `get_dummies()` can break if new categories appear.


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:4],
                                                 df.iloc[:,-1],
                                                 test_size=0.2,
                                                 random_state=2)
```

```python
x_train.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first',dtype=np.int32,sparse_output=False)
```

```python
# fit only x_train 
#  transform Both 
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
```

```python
x_test_new = ohe.transform(x_test[['fuel','owner']])
```
