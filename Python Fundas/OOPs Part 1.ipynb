### ATM Balance (Mini Project) 

```python
class Atm:

# constructor
    def __init__ (self):
        self.pin = ''
        self.balance = 0 
        self.menu()

    def menu(self):
        user_input = input('''
        HI How can I Help You 
        1.Press 1 to create pin 
        2.Press 2 to change pin 
        3.Press 3 to check balance
        4.Press 4 to withdraw
        5.Anything else to exit
        ''')

        if user_input == '1':     #create pin
            self.create_pin()
        elif user_input == '2':   #change pin 
            self.change_pin()
        elif user_input == '3':   #check balance
            self.check_balance()
        elif user_input == '4':   #withdraw
            self.withdraw()
        else:
            exit()
        
            
    def create_pin(self):
        user_pin = int(input('Enter your PIN :')) 
        self.pin = user_pin

        user_balance = int(input('Enter Your Balance:'))
        self.balance = user_balance

        print('PIN Created successfully')
        self.menu()

    def change_pin(self): 
        old_pin = int(input('Enter Your Old Pin:'))

        if old_pin == self.pin : 
             # Let the user change the pin 
            new_pin = int(input('Enter New PIN:'))
            print('PIN changed successfully')
            self.pin = new_pin
            self.menu()
        else:
            print('Nai Karne de Sakta re BABA!!')
            self.menu()
                
    def check_balance(self): 
        user_pin = int(input('Enter Your PIN :'))
        if user_pin == self.pin:
            print('Your Balance is :',self.balance)
            self.menu()
        else:
            print('Mai nhi Banatunga')
            self.menu()

    def withdraw(self):
        user_input = int(input('Enter Your PIN:'))
        if user_input == self.pin: 
        # Allow user to withdraw
            amount = int(input('Enter amount:'))
            if amount <= self.balance :
                self.balance = self.balance - amount
                print('withdrawal Successfull. Your Current Balance is :',self.balance)
            else:
                print('Aree Gareeb')
        else:
            print('Salle CHOR !')
        self.menu()
```

## Lets Make our own data Type

inbuild data types : Float,int,str

own(custom) : Fraction

```python
class Fraction:

    #parameterized Constructor (Adding more the one{self} attribute in object)
    def __init__(self,x,y):
        self.num = x
        self.den = y

    def __str__(self): #__str__ is use to return the formated output
        return '{}/{}'.format(self.num,self.den)

    # adding more Operations 
    def __add__(self,others): #(Self = fr1 , other = fr2)
        new_num = self.num*others.den + others.num*self.den
        new_den = self.den*others.den
        return '{}/{}'.format(new_num,new_den)

    def __sub__(self,others): #(Self = fr1 , other = fr2)
        new_num = self.num*others.den - others.num*self.den
        new_den = self.den*others.den
        return '{}/{}'.format(new_num,new_den)

    def __mul__(self,others): #(Self = fr1 , other = fr2)
        new_num = self.num*others.num
        new_den = self.den*others.den
        return '{}/{}'.format(new_num,new_den)

    def __truediv__(self,others): #(Self = fr1 , other = fr2)
        new_num = self.num*others.den
        new_den = self.den*others.num
        return '{}/{}'.format(new_num,new_den)

    def convert_to_deci(self):
        return self.num/self.den
```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

```python
# converting Fraction -> Decimal
fr1.convert_to_deci()
```
