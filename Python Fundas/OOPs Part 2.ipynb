### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line

```python
class Point():

    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self): 
        return '<{},{}>'.format(self.x_cod,self.y_cod) #Q1 done

    def eucliden_distance(self,other): 
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5    #Q2 Done

    def dist_from_ori(self):
        return self.eucliden_distance(Point(0,0))   #Q3 Done

class Line():

    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.A,self.B,self.C)  #Q4 Done

    def point_on_line(line,point):      #Q5 Done (IMP)
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return 'Point is On the Line'
        else:
            return 'Point is Not on the Line'

    def shortest_dist(line,point):   #Q6 (find the shortest distance ) Done
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C) / (line.A**2 + line.B**2)**0.5
```

```python
p1 = Point(0,0)
p2 = Point(-17,-1)
print(p1)
print(p2)

# p1.eucliden_distance(p2)
p2.dist_from_ori()
```

```python
l1 = Line(2,2,-8)
p1 = Point(2,2)

l1.shortest_dist(p1)
```

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
# object without a reference
class Person:

  def __init__(self):
    self.name = 'Shubham'
    self.gender = 'male'

p = Person()
q = p
# p,q,etc are just variable which holds the Object Address (IMP)
```

```python
# Multiple ref
print(id(p))
print(id(q))
```

```python
# change attribute value with the help of 2nd object
```

```python
print(p.name)
print(q.name)
q.name = 'Tina'   #This change in q will change p also , coz both poinst the same Obj address (that's Y we should handle this very carefully)
print(q.name)
print(p.name)
```

## Is Object Muteable or Immutable 

```python
class Person:

  def __init__(self,name,gender):
    self.name = name
    self.gender = gender

# outside the class -> function
def greet(person):
  person.name = 'ankit'
  return person

p = Person('nitish','male')
print('old address :',id(p))  #Calling address from the object
p1 = greet(p)
print('new address :',id(p1)) # Calling address when the object is inside variable and changed 
```

Mutability means an object’s internal state can change without changing its identity   
(i.e., its id remains the same even when accessed or modified through different references)

```python
class Atm:

  # constructor(special function)->superpower -> 
  def __init__(self):
    print(id(self))
    self.pin = ''
    self.__balance = 0         ### Private Property
    #self.menu()

  def get_balance(self):      ## Access private property
    return self.__balance

  def set_balance(self,new_value):       ## Update Property
    if type(new_value) == int:
      self.__balance = new_value
    else:
      print('beta bahot maarenge')

  def __menu(self):
    user_input = input("""
    Hi how can I help you?
    1. Press 1 to create pin
    2. Press 2 to change pin
    3. Press 3 to check balance
    4. Press 4 to withdraw
    5. Anything else to exit
    """)

    if user_input == '1':
      self.create_pin()
    elif user_input == '2':
      self.change_pin()
    elif user_input == '3':
      self.check_balance()
    elif user_input == '4':
      self.withdraw()
    else:
      exit()

  def create_pin(self):
    user_pin = input('enter your pin')
    self.pin = user_pin

    user_balance = int(input('enter balance'))
    self.__balance = user_balance

    print('pin created successfully')

  def change_pin(self):
    old_pin = input('enter old pin')

    if old_pin == self.pin:
      # let him change the pin
      new_pin = input('enter new pin')
      self.pin = new_pin
      print('pin change successful')
    else:
      print('nai karne de sakta re baba')

  def check_balance(self):
    user_pin = input('enter your pin')
    if user_pin == self.pin:
      print('your balance is ',self.__balance)
    else:
      print('chal nikal yahan se')

  def withdraw(self):
    user_pin = input('enter the pin')
    if user_pin == self.pin:
      # allow to withdraw
      amount = int(input('enter the amount'))
      if amount <= self.__balance:
        self.__balance = self.__balance - amount
        print('withdrawl successful.balance is',self.__balance)
      else:
        print('abe garib')
    else:
      print('sale chor')
```

# Static VS Instance Method

##### Points to remember about static

- Static attributes are created at class level.
- Static attributes are accessed using ClassName.
- Static attributes are object independent. We can access them without creating instance (object) of the class in which they are defined.
- The value stored in static attribute is shared between all instances(objects) of the class in which the static attribute is defined.

```python
class Lion:
  __water_source="well in the circus"

  def __init__(self,name, gender):
      self.__name=name
      self.__gender=gender

  def drinks_water(self):
      print(self.__name,
      "drinks water from the",Lion.__water_source)

  @staticmethod
  def get_water_source():
      return Lion.__water_source

simba=Lion("Simba","Male")
simba.drinks_water()
print( "Water source of lions:",Lion.get_water_source())
```
