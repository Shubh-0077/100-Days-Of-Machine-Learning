### Write OOP classes to handle the following scenarios:

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line

```python
class Point():

    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self): 
        return '<{},{}>'.format(self.x_cod,self.y_cod) #Q1 done

    def eucliden_distance(self,other): 
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5    #Q2 Done

    def dist_from_ori(self):
        return self.eucliden_distance(Point(0,0))   #Q3 Done

class Line():

    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.A,self.B,self.C)  #Q4 Done

    def point_on_line(line,point):      #Q5 Done (IMP)
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return 'Point is On the Line'
        else:
            return 'Point is Not on the Line'

    def shortest_dist(line,point):   #Q6 (find the shortest distance ) Done
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C) / (line.A**2 + line.B**2)**0.5
```

```python
p1 = Point(0,0)
p2 = Point(-17,-1)
print(p1)
print(p2)

# p1.eucliden_distance(p2)
p2.dist_from_ori()
```

```python
l1 = Line(2,2,-8)
p1 = Point(2,2)

l1.shortest_dist(p1)
```

### How objects access attributes

```python
class Person:

  def __init__(self,name_input,country_input):
    self.name = name_input
    self.country = country_input

  def greet(self):
    if self.country == 'india':
      print('Namaste',self.name)
    else:
      print('Hello',self.name)
```

```python
# how to access attributes
p = Person('shubh','india')
```

```python
p.name
p.country
```

```python
# how to access methods
p.greet()
```

### Reference Variables

- Reference variables hold the objects
- We can create objects without reference variable as well
- An object can have multiple reference variables
- Assigning a new reference variable to an existing object does not create a new object

```python
# object without a reference
class Person:

  def __init__(self):
    self.name = 'Shubham'
    self.gender = 'male'

p = Person()
q = p
# p,q,etc are just variable which holds the Object Address (IMP)
```

```python
# Multiple ref
print(id(p))
print(id(q))
```

```python
# change attribute value with the help of 2nd object
```

```python
print(p.name)
print(q.name)
q.name = 'Tina'   #This change in q will change p also , coz both poinst the same Obj address (that's Y we should handle this very carefully)
print(q.name)
print(p.name)
```
