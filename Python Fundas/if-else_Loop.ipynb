```python
# Menu driven Calulator 
fnum  = int(input('Enter first number: '))
snum  = int(input('Enter second number: '))

op = input('Enter your operator: ')

if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum * snum)
else:
    print(fnum / snum)
```
